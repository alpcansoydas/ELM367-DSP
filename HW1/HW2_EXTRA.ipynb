### SORU-1

```python
import numpy as np
a = 3-4j
b = 1+2j
C = a + b
print("Real part =",C.real) #reel
print("Imaginar part =",C.imag) #sanal
print("Amplitude =",abs(C)) #genlik
print("Phase =",np.angle(C)) #faz
```
