### Grafik çizdirmek

```python
from matplotlib import pyplot as plt
n = np.arange(0,10)
x = np.cos(n*np.pi/3+np.pi/2)
plt.stem(n,x)
plt.show()
```

```python
plt.figure()
t = np.linspace(0,2/50,1000)
x = np.cos(100*np.pi*t-np.pi/5)
plt.plot(t,x)
plt.xlabel('zaman')
plt.ylabel('x(t)')
plt.show()
```

### Tek figürde birden fazla işaret çizdirmek

```python
t = np.linspace(0,2/50,1000)
x1 = np.cos(100*np.pi*t-np.pi/5)
x2 = 2*np.sin(200*np.pi*t)
plt.plot(t,x1,t,x2)
plt.legend(('x1(t)','x2(t)'))
plt.show()
```

### SORU-1

```python
import numpy as np
a = 3-4j
b = 1+2j
D = a / b
print("Real part =",D.real) #reel
print("Imaginar part =",D.imag) #sanal
print("Amplitude =",abs(D)) #genlik
print("Phase =",np.angle(D)) #faz
```
