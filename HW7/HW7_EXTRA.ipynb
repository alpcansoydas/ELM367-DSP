#### Type-1

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sgnl
from scipy.fftpack import fft, ifft

%matplotlib notebook
import zplane

hn=[1,1,1,1,1] #Type-1 ==> filter order even and h[n] is symmetric
N=100;
w=np.linspace(0,2*np.pi,N)
Hw=np.fft.fftshift(fft(hn,N))
Hw_mag=abs(Hw)
Hw_phs=np.unwrap(np.angle(Hw))


plt.figure()
plt.plot(w/np.pi,Hw_mag) 
plt.title('$|H(\omega)|$')
plt.xlabel('$\omega$ / $\pi$') 
plt.ylabel('$H(\omega)$') 

plt.figure()
plt.plot(w/np.pi,Hw_phs)
plt.title("$ \measuredangle H(\omega)$")
plt.xlabel('$\omega$ / $\pi$')
plt.ylabel('$H(\omega)$') 

#group delay
tau_w=-np.diff(Hw_phs)*(N/(2*np.pi))

plt.figure() 
plt.plot(w[1:]/np.pi,tau_w) 
plt.title('$H(\omega) $ - grup gecikmesi')
plt.xlabel('$\omega$ / $\pi$') 
plt.ylabel('$H(\omega)$') 

#z plane
num = hn
denum = np.array([1])
zplane.zplane(num,denum)
```

#### Type-4

```python
hn=[1,1,-1,-1] # Type-4 => Filter order odd and h[n] is anti-symmetric
N=100;
w=np.linspace(0,2*np.pi,N) #0,2*pi aralaığı belirlenmesi
Hw=np.fft.fftshift(fft(hn,N)) #Fourier dönüşümü alınması
Hw_mag=abs(Hw) #fft si alınan fonksiyonun genliği
Hw_phs=np.unwrap(np.angle(Hw)) #sıkıştırlmamış faz cevabı için unwrap kullanıldı

# genlik ve faz grafiklerinin çizdirilmesi
plt.figure()
plt.plot(w/np.pi,Hw_mag) 
plt.title('$|H(\omega)|$')
plt.xlabel('$\omega$ / $\pi$') 
plt.ylabel('$H(\omega)$') 

plt.figure()
plt.plot(w/np.pi,Hw_phs)
plt.title("$ \measuredangle H(\omega)$")
plt.xlabel('$\omega$ / $\pi$')
plt.ylabel('$H(\omega)$') 

#grup gecikmesi grafiğinin çizdirilmesi
tau_w=-np.diff(Hw_phs)*(N/(2*np.pi)) # faz cevabının türevinin negatifi grup gecikmesidir.

plt.figure() 
plt.plot(w[1:]/np.pi,tau_w) 
plt.title('$H(\omega) $ - grup gecikmesi')
plt.xlabel('$\omega$ / $\pi$') 
plt.ylabel('$H(\omega)$') 

#z ekseni kutup diagarmının çizdirlmesi
num = hn
denum = np.array([1])
zplane.zplane(num,denum)
```
